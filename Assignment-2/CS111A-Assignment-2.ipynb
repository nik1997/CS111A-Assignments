```python
### B.2.a

y(x) = x ^ (1/3)
show(y)

y_prime(x) = diff(y, x)

x0 = 8
L(x) = y(x0) + y_prime(x0)*(x - x0)
show(L)

plot_y = plot(y, (x, 0, 15), color='blue')
plot_L = plot(L, (x, 0, 15), color='red')

x_10(y) = 10
# plot_10 = plot(l,  color='green', linestyle='--' )
plot_10 = line([[10,0],[10,3]], color='green', linestyle='--')
plot_y + plot_L + plot_10
```
