```python
# A.2.a
c = 3*10^8 #speed of light
L0 = 111 #length of rocket in m Length of Saturn V rocket

L(v) = L0*sqrt(1 - (v^2/c^2))
show(L)
plot(L, (v, 0, 3*10^8))
```

```python
# A.2.b As v increases, L gets smaller and smaller
```

```python
# A.2.c
print(limit(L, v = c, dir='-'))
plot(L, (v, 0, 3*10^8), axes_labels=['Length of rocket appearing to observer (m)','Speed of the rocket (m/s)'])

# As the v tends toward c, the length of the rocket tends towards 0.
# Thus, ,the object will not be seen.
# There is nothing greater than speed of light and v cannot be greater than c else the square root cannot have negative number
```

```python
# A.2.d
# The left-hand limit was needed, because it denotes that the rocket is
# going in the opposite direction of light or going further away.
# Thus, as it goes further away the object seems to vanish
```

```python
C = plot(ceil(x), (x, -5, 5), color='blue')
F = plot(floor(x), (x, -5, 5), color='green')
S = plot(sgn(x), (x, -5, 5), color='red')
show(C)
show(F)
show(S)
```

a) Sgn is not continous at x=0, floor and ceiling at x = C, where C is any integer
b)
    Ceil -> To create a upperbound for the maximum allowed current in any electrical fuse. If we need to buy 20 beers and they are sold at a package of 6 beers we need to buy ceil(20/6) beers
    Floor -> Dividing a group of objects equally and being able to hold the extra stuffs. For example, dividing 20 crayons to 6 kids equally, we would need floor(30/6) = 3 crayons each. So that the kids do not fight over it.
    Sgn -> Used when we only need to direction or need boolean value. For instance, switching on the heater when it crosses a certain threshold forinstance. f(x) = sgn(20 - T) and our threshold  is 20C, so if the temperature goes below, it gives a positive value that is the signal to turn on the heater.
