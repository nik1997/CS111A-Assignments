## A.1

```python
var('Y, L, K')

Y(L, K) = 2 * L^(0.75) * K^(0.25)
show(Y)

# Y = total value of goods
# L = total person hour worked
# K = value of capital input
print Y(1000, 1000000)
```

```python
Y.diff()
show(Y.diff())

# equn of tangent plane => z = z0 + dz/dx(x-x0) + dz/dy(y-y0)

L0 = 1000
K0 = 1000000
Y0 = Y(L0, K0)

print(L0, K0, Y0)
print(Y0)
T(L, K) = Y0 + (Y.diff()[0](L0, K0)*(L - L0)) + (Y.diff()[1](L0,K0)*(K - K0))
print "Equation of tangent"
show(T)

print(Y.diff()[0](L0, K0), Y.diff()[1](L0, K0))
print(Y.diff()[0](L0, K0) * L0)
print(Y.diff()[1](L0, K0) * K0)
print(Y0 - Y.diff()[0](L0, K0) * L0 - Y.diff()[1](L0, K0) * K0)

print "T(1001,1000001) = " , T(1001,1000001)
print "T(1001,1000001) - T(1000,1000000) = " , T(1001,1000001) - T(1000,1000000)
print "Y(1001, 1000001) - Y(1000, 1000000) = ", Y(1001, 1000001) - Y(1000, 1000000)
```

## A.2

```python
var('x,y,z')
T(x,y,z) = 360/sqrt(x^2 + y^2 + z^2)
show(T)

show(T.diff())

Duf(x,y,z) = ((1/sqrt(3))*T.diff()[0]) - ((1/sqrt(3))*T.diff()[1]) - ((1/sqrt(3))*T.diff()[2])
show(Duf)
print 'Rate of change from (1,2,2)', Duf(1,2,2)
```
